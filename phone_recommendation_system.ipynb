```python
clean_names = []

for i in df['Name']:
    if "(" in i:
        ind = i.find('(')
        clean_text = i[:ind].strip()
    else:
        clean_text = i.strip()

    clean_names.append(clean_text)

df['Name'] = clean_names
```

```python
df['processor'] = df['processor'].str.replace(",","").str.replace("Processor","")
```

```python
df['storage'] = df['storage'].str.replace(",","").str.replace('inbuilt',"")
df['storage'] = df['storage'].str.replace(r'\s*GB', 'GB', regex=True)
```

```python
df['battery'] = df['battery'].str.split('Battery').str[0]
df['battery'] = df['battery'].str.replace(' ', '').str.replace(' ', '')
```

```python
df['camera'] = df['camera'].str.split('&').str[1]
```

```python
df['Combined'] = df['Name'] + " " + df['processor'] + " " + df["storage"] + " "+ df['battery'] + " " + df['camera'] + " " + df['version']
```

```python
df['Combined'].isnull().sum()
```

```python
df['Combined'] = df['Combined'].fillna("")
```

```python
df['Combined']
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['Combined'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# cosine_similarity = cosine_similarity(tfidf_matrix)
```

```python
def recommend(user_input):
    tfidf_uip = tfidf.transform([user_input])
    similarity = cosine_similarity(tfidf_uip,tfidf_matrix).flatten()
    top5_idx = similarity.argsort()[::-1]
    top_5 = top5_idx[:5]

    return df.iloc[top_5][['Name','price']] 
```

```python
recommend("32990 dimensity 8350 octa core 3.35 ghz	8gb ram 256gb6000mah 50 mp front camera	android v15")
```
